```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl
import pandas as pd
from collections import OrderedDict
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
url="http://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;page=4;team=6;template=results;type=bowling"
html = urllib.request.urlopen(url, context=ctx).read()
```

```python
temp_data=OrderedDict()
list_of_dict=[]
bs=BeautifulSoup(html, "lxml")
table_body=bs.find_all('tbody')
table_body
table_body_1 = table_body[0]
```

```python
table_body_1
```

```python
rows = table_body_1.find_all('tr')
rows
```

```python
for row in rows:
    cols=row.find_all('td')
    cols=[x.text.strip() for x in cols]
    temp_data=OrderedDict()
    for i in range(len(cols)):
        temp_data["Player"]=cols[0]
        temp_data["Span"]=cols[1]
        temp_data["Matches"]=cols[2]
        temp_data["Innings"]=cols[3]
        temp_data["Balls"]=cols[4]
        temp_data["Runs"]=cols[5]
        temp_data["Wkts"]=cols[6]
        temp_data["BBI"]=cols[7]
        temp_data["Ave"]=cols[8]
        temp_data["Econ"]=cols[9]
        temp_data["SR"] = cols[10]
        temp_data["4"] = cols[11]
        temp_data["5"] = cols[12]
    list_of_dict.append(temp_data)
```

```python
df=pd.DataFrame(list_of_dict)
```

```python
df
```

```python
writer = ExcelWriter('ip_b4.xlsx')
df.to_excel(writer,'Sheet1',index=False)
writer.save()
```
